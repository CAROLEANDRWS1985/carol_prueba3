```python
from datetime import date

# Variables globales
pisos = 10
departamentos_por_piso = 4
precios = {'A': 3800, 'B': 3000, 'C': 2800, 'D': 3500}
departamentos_disponibles = [[True for _ in range(departamentos_por_piso)] for _ in range(pisos)]
compradores = {}

# Función para mostrar el menú y obtener la opción seleccionada
def mostrar_menu():
    print("\n-- Menú --")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")
    opcion = input("Ingrese una opción: ")
    return opcion

# Función para mostrar el estado de los departamentos
def mostrar_estado_departamentos():
    print("\n-- Estado de los departamentos --")
    for piso in range(pisos):
        print(f"Piso {piso+1}: ", end="")
        for i, disponible in enumerate(departamentos_disponibles[piso]):
            departamento = chr(65+i)  # Convertir número a letra (65 es el código ASCII de 'A')
            estado = "Disponible" if disponible else "Vendido"
            print(f"{departamento}{piso+1}: {estado}", end="  ")
        print()

# Función para comprar un departamento
def comprar_departamento():
    mostrar_estado_departamentos()
    departamento_valido = False
    while not departamento_valido:
        piso = int(input("Ingrese el número de piso: "))
        tipo = input("Ingrese el tipo de departamento (A, B, C o D): ").upper()
        departamento_index = ord(tipo) - 65  # Convertir letra a número (65 es el código ASCII de 'A')
        if piso < 1 or piso > pisos or departamento_index < 0 or departamento_index >= departamentos_por_piso:
            print("Departamento inválido. Intente nuevamente.")
        elif not departamentos_disponibles[piso-1][departamento_index]:
            print("Departamento no disponible. Intente nuevamente.")
        else:
            departamento_valido = True
            departamentos_disponibles[piso-1][departamento_index] = False
            run = input("Ingrese el RUN del comprador (sin guion ni puntos): ")
            compradores[run] = f"{tipo}{piso}"
            print("Operación realizada correctamente.")

# Función para mostrar el listado de compradores
def mostrar_listado_compradores():
    print("\n-- Listado de compradores --")
    for run in sorted(compradores):
        print(f"RUN: {run}, Departamento: {compradores[run]}")

# Función para mostrar las ganancias totales
def mostrar_ganancias_totales():
    total_por_tipo = {tipo: 0 for tipo in precios}
    total_general = 0
    for piso in range(pisos):
        for i, vendido in enumerate(departamentos_disponibles[piso]):
            if not vendido:
                departamento = chr(65+i)
                tipo = departamento
                precio = precios[tipo]
                total_por_tipo[tipo] += precio
                total_general += precio
    print("\n-- Ganancias totales --")
    print("Tipo de Departamento   Cantidad   Total")
    for tipo, total in total_por_tipo.items():
        cantidad = total // precios[tipo]
        print(f"{tipo} {' '*(21-len(tipo))} {cantidad} {' '*(10-len(str(cantidad)))} {total} UF")
    print(f"\nTOTAL {' '*(21-5)} {sum(total_por_tipo.values())} UF")

# Función principal
def main():
    while True:
        opcion = mostrar_menu()

        if opcion == '1':
            comprar_departamento()
        elif opcion == '2':
            mostrar_estado_departamentos()
        elif opcion == '3':
            mostrar_listado_compradores()
        elif opcion == '4':
            mostrar_ganancias_totales()
        elif opcion == '5':
            nombre = input("Ingrese su nombre: ")
            apellido = input("Ingrese su apellido: ")
            fecha_actual = date.today().strftime("%d/%m/%Y")
            print(f"\nGracias por utilizar el sistema, {nombre} {apellido}.")
            print(f"Sesión finalizada el {fecha_actual}.")
            break
        else:
            print("Opción inválida. Intente nuevamente.")

# Ejecutar la función principal
main()
```
